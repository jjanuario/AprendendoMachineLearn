```python
#Vamos carregar p arquivo com o resultado da eleção no Brasil dividifo por monucipio/zona
df_votacao = pd.read_csv('br_municipios/votacao_candidato_munzona_2018_BR.csv', sep=';', encoding='latin1')
```

```python
df_votacao.head().T
```

```python
#Separar os dados pois precisamos apenas dos dados de 2 Turno
df_2turno = df_votacao[df_votacao['NR_TURNO']==2]
```

```python
df_2turno.sample(3)
```

```python
#Vamos verificar quantos municipios unicos existem no conjunto de dados
# Vamos pegar por codigo do municipio
df_2turno['CD_MUNICIPIO'].nunique()
```

```python
# Mas o Brasil  tem 5570 municipios, como pode nosso dataframe ter 5741 são 171 municipios a mais.

# Exibindo algumas cidades fora do Brasil.
df_2turno[df_2turno['SG_UF']=='ZZ'].sample(4)
```

```python
df_2turno[df_2turno['SG_UF'] == 'ZZ']['CD_MUNICIPIO'].nunique()
```

```python
#Vamos eleminar os municipios de fora do Brasil
df_2turno = df_2turno[df_2turno['SG_UF'] != 'ZZ']
```

```python
# Os dados estão separados por municipios / zona
# municipios pequenos possuem apenas uma zona eleitoral
# mas municipios maiores podem ter varias zonas eleitorais

# Vamos verificar Belo Horizonte
df_2turno[df_2turno['NM_MUNICIPIO'] == 'SÃO PAULO']
```

```python
#Eliminar algumas colunas do dataframe que não serão utilizadas
df_2turno_limpo = df_2turno[['SG_UF','CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]
df_2turno_limpo.head()
```

```python
# Precisamos agrupar os dados por municipio e candidatos
df_2turno_limpo.groupby(['SG_UF', 'CD_MUNICIPIO','NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# Fazer groupby, reindexar o indice, vai salvar isso em um dataframe

df_2turno_limpo_reindexado = df_2turno.groupby(['SG_UF', 'CD_MUNICIPIO','NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()
df_2turno_limpo_reindexado.sample(4)
```

```python
#Ordenar o dataframe pela quantidade de votos e depois aoenas o primiero 
# registro de cada municipio
df_2turno_limpo_reindexado.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)
```

```python
df_2turno_limpo_reindexado.head(10)
```

```python
#dropas as linhas duplicadas, mantendo a primeira
df_2turno_limpo_reindexado.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)
```

```python
df_2turno_limpo_reindexado.head(10)
```

```python
# Agora vamos criar uma cooluna representando o resultado da eleicao naquele municipio
#sendo que o resultado é True se Bolsonaro e False se Haddad
df_2turno_limpo_reindexado['RESULTADO'] = df_2turno_limpo_reindexado['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
df_2turno_limpo_reindexado.head(10)
```

```python
#Mostrando valores do dataframe geopandas
df_mapa.sample(10)
```

```python
#tenrando encontrear municipoo pelo codifo
df_2turno_limpo_reindexado['CD_MUNICIPIO'] == 133]
```

```python
#Porque não podemos usar o nomedo municipio para juntar 
df_2turno_limpo_reindexado[df_2turno_limpo_reindexado['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
#Aqui pego o dataframe de equivalencia entre TSE e IBGE para utilizar os codigos de ambos
df_equivalencia = pd.read_csv('data/correspondencia-tse-ibge.csv')
```

```python
# Vizualizando o dataframe
df_equivalencia.head(10)
```

```python
# Vamos preparar os dataframes para plotar o mapar do Brasil vai da tudo vende!
#Vamos reindexar o dataframe de votação pelo codigo do TSE
df_vot_novo = df_2turno_limpo_reindexado.copy()
df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)
df_vot_novo.head(10)
```

```python
#Vamos tambem reindexar o dataframe de equivalencia pelo codigo do TSE
df_equi_novo = df_equivalencia.copy()
df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)
df_equi_novo.head()
```

```python
#Verificando o tamanho dos dataframes
df_vot_novo.info()
```

```python
#Verificando o tamanho dos dataframes
df_equi_novo.info()
```

```python
# Criando um novo dataframe com base no merge de votação e equivalencia
df_vot_equi = df_vot_novo.join(df_equi_novo)
df_vot_equi.head()
```

```python
#Vamos juntar o datarframe do GEOPANDAS com o dataframe de votação e equivalencia
df_mapa.info()
```

```python
#transformar a coluna de codigo IBGE do dataframe de votação / Equivalencvia
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)
df_vot_equi.info()
```

```python
#Vamos unificar o dataframe do Mapa com o dataframe 
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)
```

```python
#Mudando o indice do dataframe de votação /equivalencia
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
#vamos juntas os dataframe
df_mapa_novo = df_mapa.join(df_vot_equi)
df_mapa_novo.head(10).T
```

```python

#Verificando o tamonho do mapa novo
df_mapa_novo.info()
```

```python
#Vamos descobrir os dois municipios a mais no mapa
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
#Matando a coluna de ajuste
df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
#Vamos eliminar a informação das duas Lagoas do RS
df_mapa_novo.dropna(inplace=True)
```

```python
#Vamos desenhar o mapa
#Criar a referencia para o codigo de cor (false/0 = red . True/1 = green)
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['red','green'])

#Desenhand o mapa
fig, ax = plt.subplots(1, figsize=(12,12))
ax = df_mapa_novo.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.1, ax=ax, edgecolor='grey')
#Escondendo o eixos
ax.set_axis_off()

#Colocando um titulo
ax.set_title('Resultado Eleições 2018 por municipio / 2º Turno', fontdict={'fontsize':'25','fontweight': '3'})

#Criando uma nota de redapé
ax.annotate('Fonte: IBGE. 2015; TSE, 2019; Jornal O Estado de São Paulo, 2019',
            xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
            verticalalignment='top', fontsize=10, color='#555555')
#Alterando a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')
plt.show()
```
